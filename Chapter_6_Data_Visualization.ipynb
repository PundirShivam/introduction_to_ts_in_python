# Time Series Visualization
1. Line Plots.
2. Histograms and Density Plots.
3. Box and Whisker Plots.
4. Heat Maps.
5. Lag Plots or Scatter Plots.
6. Autocorrelation Plots.

```python
temp = pd.read_csv(r'./data/daily-min-temperatures.txt',index_col='Date',parse_dates=True)
```

# Line plot

```python
temp.plot(grid=True,legend='').set(title="Daily Min Temperature",xlabel='',ylabel='')
```

```python
(temp.plot(grid=True,legend='',style='.',color='k').
    set(title="Daily Min Temperature",xlabel='',ylabel=''))
```

```python
# subplots for each year
import matplotlib.pyplot as plt
grouped = temp.groupby(pd.Grouper(freq='Y')) # yearlly
temp_yr = pd.DataFrame()
for yr,grp in grouped:
    temp_yr[yr.year] = grp['Temp'].values
_ = temp_yr.plot(subplots=True,figsize=(12,8),grid=True)
plt.tight_layout(True)
```

# Histogram

```python
_ = temp.hist(bins=20)
_ = plt.title("Histogram of Daily minimum temparature")
```

```python
_ = temp.plot(kind='kde',grid=True)
_ = plt.title("Density Plot of Daily minimum temparature")
```

# Box and Whisker Plot

```python
# 25th percentile - lowest line
# 75th percentile - highest line
# 50th percentile - middle line 
temp_yr.boxplot(figsize=(8,6))
```

```python
grouped = temp[temp.index.year==1990].groupby(pd.Grouper(freq='M'))
month_list = []
grp_list = []
for mnth, grp in grouped:
    grp_list.append(pd.Series(grp['Temp'].values,name=mnth.month))
temp_month = pd.concat(grp_list,axis=1)
temp_month.boxplot()
_ = plt.title("Box plot for year 1990")
```

# Heat Maps

```python
plt.matshow(temp_yr.T,aspect='auto')
```

```python
plt.matshow(temp_month.T,aspect='auto',interpolation='None')
```

# Lag - Scatter plots

```python
pd.plotting.lag_plot(temp)
plt.grid(True)
_ = plt.title("Lag-Scatter Plot")
```

```python
f , ax = plt.subplots(4,3,figsize=(12,12))
ax = ax.flatten()
for i in range(1,13):
    pd.plotting.lag_plot(temp,lag=i,ax=ax[i-1]).set(title="t vs t-"+str(i))
    plt.tight_layout(True)
```

# Autocorrelation Plots

```python
_ = pd.plotting.autocorrelation_plot(temp['Temp'])
_ = plt.title("Autocorrelation plot")
```
